```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.iloc[:,[3,4]].values
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters = i ,init = 'k-means++',random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('the elbow method')
plt.xlabel('number of clusters')
plt.ylabel('wcss')
plt.show()
```

```python
 kmeans = KMeans(n_clusters = 5 ,init = 'k-means++',random_state=42)
 y_means =   kmeans.fit_predict(X)
```

```python
print(y_means)
```

```python
plt.scatter(X[y_means==0,0],X[y_means==0,1],s=80,c='red',label='cluster1')
plt.scatter(X[y_means==1,0],X[y_means==1,1],s=80,c='blue',label='cluster2')
plt.scatter(X[y_means==2,0],X[y_means==2,1],s=80,c='green',label='cluster3')
plt.scatter(X[y_means==3,0],X[y_means==3,1],s=80,c='cyan',label='cluster4')
plt.scatter(X[y_means==4,0],X[y_means==4,1],s=80,c='magenta',label='cluster5')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')
plt.title('cluster of customers')
plt.xlabel('annual income')
plt.ylabel('spending score')
plt.legend()
plt.show()
```
