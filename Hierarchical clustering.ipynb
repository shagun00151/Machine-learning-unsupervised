```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.iloc[:,[3,4]].values
```

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X,method='ward'))
plt.title('Dendrograms')
plt.xlabel('Customers')
plt.ylabel('Euclidean distance')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
model = AgglomerativeClustering(n_clusters = 5,affinity='euclidean',linkage='ward')
y_model = model.fit_predict(X)
```

```python
print(y_model)
```

```python
plt.scatter(X[y_model==0,0],X[y_model==0,1],s=100,c='red',label='cluster1')
plt.scatter(X[y_model==1,0],X[y_model==1,1],s=100,c='blue',label='cluster2')
plt.scatter(X[y_model==2,0],X[y_model==2,1],s=100,c='green',label='cluster3')
plt.scatter(X[y_model==3,0],X[y_model==3,1],s=100,c='cyan',label='cluster4')
plt.scatter(X[y_model==4,0],X[y_model==4,1],s=100,c='magenta',label='cluster5')
plt.title('cluster of customers')
plt.xlabel('annual income')
plt.ylabel('spending score')
plt.legend()
plt.show()
```
